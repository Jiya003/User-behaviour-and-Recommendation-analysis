# Outlier Analysis

## Z-Score Method

```python
from scipy import stats
import numpy as np

# Calculate Z-scores
z_scores = np.abs(stats.zscore(data[['Quantity', 'UnitPrice']]))
threshold = 3  # Common threshold for identifying outliers

# Identify outliers
outliers_z = (z_scores > threshold)

# Display the number of outliers found for each column
num_outliers_z = outliers_z.sum(axis=0)
print(f"Number of outliers detected using Z-score:\n{num_outliers_z}")
```

## IQR Method (BoxPlot)

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1_quantity = data['Quantity'].quantile(0.25)
Q3_quantity = data['Quantity'].quantile(0.75)
IQR_quantity = Q3_quantity - Q1_quantity

Q1_unit_price = data['UnitPrice'].quantile(0.25)
Q3_unit_price = data['UnitPrice'].quantile(0.75)
IQR_unit_price = Q3_unit_price - Q1_unit_price

# Define lower and upper bounds for Quantity
lower_bound_quantity = Q1_quantity - 1.5 * IQR_quantity
upper_bound_quantity = Q3_quantity + 1.5 * IQR_quantity

# Define lower and upper bounds for Unit Price
lower_bound_unit_price = Q1_unit_price - 1.5 * IQR_unit_price
upper_bound_unit_price = Q3_unit_price + 1.5 * IQR_unit_price

# Display the bounds
print("Quantity Bounds: ", lower_bound_quantity, upper_bound_quantity)
print("Unit Price Bounds: ", lower_bound_unit_price, upper_bound_unit_price)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style
sns.set(style="whitegrid")

# Create a figure for boxplots
plt.figure(figsize=(12, 6))

# Boxplot for Quantity
plt.subplot(1, 2, 1)
sns.boxplot(x=data['Quantity'])
plt.title('Boxplot of Quantity (Including Outliers)')

# Boxplot for Unit Price
plt.subplot(1, 2, 2)
sns.boxplot(x=data['UnitPrice'])
plt.title('Boxplot of Unit Price (Including Outliers)')

# Adjust layout
plt.tight_layout()
plt.show()
```

## Removing Outliers

```python
# Removing outliers based on the IQR method
data_no_outliers = data[~((data['Quantity'] < lower_bound_quantity) |
                           (data['Quantity'] > upper_bound_quantity) |
                           (data['UnitPrice'] < lower_bound_unit_price) |
                           (data['UnitPrice'] > upper_bound_unit_price))]

# Verify the shape of the new DataFrame
print(f"Original data shape: {data.shape}")
print(f"Data shape after removing outliers: {data_no_outliers.shape}")
```

# APIORI ANALYSIS

```python
# Sample 10% of the DataFrame
sampled_df = df.sample(frac=0.1, random_state=42)  # Set random_state for reproducibility
```

Creating transactional data

```python
# Create a list of transactions
transactions = sampled_df.groupby('CustomerID')['Description'].apply(list).reset_index(name='Products')
```

## Encoding the data

```python
from mlxtend.preprocessing import TransactionEncoder

# Fit the transaction encoder
encoder = TransactionEncoder()
onehot = encoder.fit(transactions['Products']).transform(transactions['Products'])
df_onehot = pd.DataFrame(onehot, columns=encoder.columns_)
```

## Applying

```python
from mlxtend.frequent_patterns import apriori

# Apply Apriori
frequent_itemsets = apriori(df_onehot, min_support=0.005, use_colnames=True)  # Adjust min_support as needed
```

```python
print("Frequent Itemsets:")
print(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules

# Generate association rules with a suitable minimum threshold
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)  # You can adjust the min_threshold if needed

# Check the rules DataFrame
print("Generated Rules:")
print(rules.head())  # Display the first few rows of the rules DataFrame
```

```python
# Display the top rules
rules.sort_values(by='lift', ascending=False).head(10)
```

```python
sampled_df.head()
```

```python
print("Rules DataFrame Structure:")
print(rules.head())  # Show the first few rows of the rules DataFrame
print("Columns in Rules DataFrame:")
print(rules.columns)  # List the columns present in the DataFrame
```

```python
def get_recommendations(product_name):
    # Filter the rules for the given product name
    product_rules = rules[rules['antecedents'].apply(lambda x: product_name in x)]

    # Extract the top 5 recommendations based on lift
    recommendations = product_rules[['consequents', 'support', 'confidence', 'lift']].nlargest(5, 'lift')

    return recommendations

# Example usage
product_to_recommend = 'HEART OF WICKER LARGE'  # Replace with your product name
recommendations = get_recommendations(product_to_recommend)
print(recommendations)
```
