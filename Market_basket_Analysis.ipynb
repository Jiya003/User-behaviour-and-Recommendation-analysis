# Outlier Analysis

## Z-Score Method

```python
from scipy import stats
import numpy as np

# Calculating the Z-scores
z_scores = np.abs(stats.zscore(data[['Quantity', 'UnitPrice']]))
threshold = 3  # Common threshold for identifying outliers

# Identifying outliers, using threshold
outliers_z = (z_scores > threshold)

# Displaying the number of outliers found for each column
num_outliers_z = outliers_z.sum(axis=0)
print(f"Number of outliers detected using Z-score:\n{num_outliers_z}")
```

## IQR Method (BoxPlot)

```python
# Calculating the Q1 (25th percentile) and Q3 (75th percentile)
Q1_quantity = data['Quantity'].quantile(0.25)
Q3_quantity = data['Quantity'].quantile(0.75)
IQR_quantity = Q3_quantity - Q1_quantity

Q1_unit_price = data['UnitPrice'].quantile(0.25)
Q3_unit_price = data['UnitPrice'].quantile(0.75)
IQR_unit_price = Q3_unit_price - Q1_unit_price

# Defining the lower and upper bounds for Quantity
lower_bound_quantity = Q1_quantity - 1.5 * IQR_quantity
upper_bound_quantity = Q3_quantity + 1.5 * IQR_quantity

# Same for the Unit Price
lower_bound_unit_price = Q1_unit_price - 1.5 * IQR_unit_price
upper_bound_unit_price = Q3_unit_price + 1.5 * IQR_unit_price

# Displaying the bounds
print("Quantity Bounds: ", lower_bound_quantity, upper_bound_quantity)
print("Unit Price Bounds: ", lower_bound_unit_price, upper_bound_unit_price)
```

# FP Growth ANALYSIS

```python
# Sample 10% of the data from the DataFrame
sampled_df = df.sample(frac=0.1, random_state=42)  # Set random_state for reproducibility
```

Creating transactional data

```python
# Creating a list of transactions
transactions = sampled_df.groupby('CustomerID')['Description'].apply(list).reset_index(name='Products')
```

```python
sampled_df.shape
```

## Encoding the data

```python
from mlxtend.preprocessing import TransactionEncoder

# Fitting the transaction encoder
encoder = TransactionEncoder()
onehot = encoder.fit(transactions['Products']).transform(transactions['Products'])
df_onehot = pd.DataFrame(onehot, columns=encoder.columns_)
```

```python
df_onehot.shape
```

```python
df_onehot.head()
```

## Applying

```python
from mlxtend.frequent_patterns import apriori

# Applying Apriori Algo
frequent_itemsets = apriori(df_onehot, min_support=0.005, use_colnames=True)
```

```python
print("Frequent Itemsets:")
print(frequent_itemsets.tail())
```

```python
print(frequent_itemsets.head())
print(frequent_itemsets.info())
```

```python

print(frequent_itemsets.dtypes)
```

```python
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
import time
start_time=time.time()
# Apply FP-Growth Algorithm to find frequent itemsets
frequent_itemsets = fpgrowth(df_onehot, min_support=0.005, use_colnames=True)

# Print the frequent itemsets
print("Frequent Itemsets Using FP-Growth Algorithm:\n", frequent_itemsets)

# **Determine num_itemsets - for example, using the length of the frequent itemsets:**
num_itemsets = len(frequent_itemsets)

# Generate association rules from the frequent itemsets, including num_itemsets
rules = association_rules(frequent_itemsets, num_itemsets, metric="lift", min_threshold=1.0)

# Display the rules
print("\nGenerated Association Rules from FP-Growth:\n", rules)
end_time = time.time()
print(f"Time taken for FP: {end_time - start_time} seconds")
```

```python
# Displaying the top rules
rules.sort_values(by='lift', ascending=False).head(10)
```

```python
print("Rules DataFrame Structure:")
print(rules.head())  # for showing the first few rows of the rules DataFrame
print("Columns in Rules DataFrame:")
print(rules.columns)  # To List the columns present in the DataFrame
```

```python
def get_recommendations(product_name):
    # Filtering out the rules for the given product name
    product_rules = rules[rules['antecedents'].apply(lambda x: product_name in x)]

    # For Extracting the top 5 recommendations based on lift
    recommendations = product_rules[['consequents', 'support', 'confidence', 'lift']].nlargest(5, 'lift')

    return recommendations

product_to_recommend = 'LUNCH BAG RED RETROSPOT'
recommendations = get_recommendations(product_to_recommend)
print(recommendations)
```

# Apriori analysis

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import time

# Starting the timer
start_time = time.time()

# Handling NaN values by filling the missing values (example: fill with 0)
sampled_df_cleaned = sampled_df.fillna(0)
sampled_df_cleaned = sampled_df_cleaned.sample(frac=0.1, random_state=42)

# Converting the relevant columns to binary (1 if purchased, 0 if not)

unique_products = sampled_df_cleaned['StockCode'].unique()


transaction_matrix = pd.DataFrame(0, index=sampled_df_cleaned.index, columns=unique_products)


for index, row in sampled_df_cleaned.iterrows():
    transaction_matrix.loc[index, row['StockCode']] = 1



frequent_itemsets_1 = apriori(transaction_matrix, min_support=0.001, use_colnames=True)
num_itemsets=len(frequent_itemsets_1)

if frequent_itemsets_1.empty:
    print("No frequent itemsets found. Consider lowering the min_support value.")
else:
    # Generating the association rules from frequent itemsets
    rules_1 = association_rules(frequent_itemsets_1,num_itemsets, metric="confidence", min_threshold=0.005)

    # Stopping the timer
    end_time = time.time()

    # Calculating the total time taken
    execution_time = end_time - start_time

    # Displaying the results and time taken
    print("Frequent Itemsets:\n", frequent_itemsets_1)
    print("\nAssociation Rules:\n", rules_1)
    print(f"\nTime taken for Apriori and rule generation: {execution_time:.4f} seconds")
```

# **Analysis:**

In my experiment, I compared the execution times of **FP-Growth** and **Apriori** algorithms on different portions of the dataset. The results were as follows:

1. **FP-Growth (0.1% of data, 1.55 sec)**:
   - FP-Growth was applied to 0.1% of the dataset and completed in 1.55 seconds. This shows that FP-Growth can efficiently mine frequent itemsets even when dealing with a larger sample of the data. The algorithm's ability to work with compressed data (using the FP-tree structure) allows it to perform faster, even as the dataset size increases.

2. **Apriori (0.01% of data, 5.22 sec)**:
   - Apriori was tested on a much smaller sample (0.01% of the data), yet it still took 5.22 seconds to complete. This indicates that **Apriori's performance degrades significantly**, even when a smaller portion of the data is used.
